```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


models = ['Random Forest', 'XGBoost', 'SVC']
metrics = ['Accuracy', 'Precision', 'Recall', 'ROC AUC', 'F1 Score']
values = np.array([
    [0.9703,0.9788,0.9833,0.9894,0.9808],
    [0.9728,0.9673,0.9989,0.9841,0.9827],
    [0.9772,0.9748,0.9963,0.9876,0.9854]
])

colors = ['blue', 'RED', 'green']
bar_width = 0.2
index = np.arange(len(metrics))

# Plotting each metric for each model
plt.figure(figsize=(10, 6))

for i, model in enumerate(models):
    plt.bar(index + i * bar_width, values[i], bar_width, label=model, color=colors[i])

# plt.figure(figsize=(10, 6))
# Generating y-axis tick values with a step of 0.1
tick_values = np.arange(0, 1 + 0.1, 0.1)
plt.yticks(tick_values)

plt.xlabel('Metrics')
plt.ylabel('Values')
plt.title('Comparison of Metrics for Different Models after 10 iterations of 5-fold cross-validation')
plt.xticks(index + bar_width * len(models) / 2, metrics)  # Adjust x-axis ticks

# Show legend at a higher position
# plt.legend()# 
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), shadow=True, ncol=len(models))

plt.tight_layout()
plt.show()
```
