```python
import matplotlib.pyplot as plt
import numpy as np

# Example metrics for three models (replace these with your actual metrics)
models = ['Random Forest', 'XGBoost', 'SVC']
metrics = ['Accuracy', 'Precision', 'ROC AUC', 'Recall', 'F1 Score']
values = np.array([
    [0.93340943,0.93063633,0.9926494,0.989411,0.95889],
    [0.95559360,0.952456831,0.988060129,0.99298245,0.97210341],
    [0.8227929,0.81489400,0.5732602,1,0.8978577]
])

colors = ['blue', 'RED', 'green']
bar_width = 0.2
index = np.arange(len(metrics))

# Plotting each metric for each model
plt.figure(figsize=(10, 6))

for i, model in enumerate(models):
    plt.bar(index + i * bar_width, values[i], bar_width, label=model, color=colors[i])

# plt.figure(figsize=(10, 6))

plt.xlabel('Metrics')
plt.ylabel('Values')
plt.title('Comparison of Metrics for Different Models after 5-fold cross-validation')
plt.xticks(index + bar_width * len(models) / 2, metrics)  # Adjust x-axis ticks

# Show legend at a higher position
# plt.legend()# 
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), shadow=True, ncol=len(models))

plt.tight_layout()
plt.show()
```
