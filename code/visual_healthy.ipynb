```python
test_load_label = nib.load('../dataset/Healthy_Pancreas/TCIA_pancreas_labels-02-05-2017/label0015.nii.gz').get_fdata()
test_load_label.shape
slices = test_load_label.shape[2]
```

```python
slices
```

```python
def find_slice(slices,category):
    best = 0
    slice_no = 0
    for x in range(slices):
        mat = np.matrix(test_load_label[:, :, x])
        current = np.count_nonzero(mat == category)

        if current > best:
           
            best = current
            slice_no = x
    return slice_no
```

```python
pancreas_best = find_slice(slices,1)
print(pancreas_best)
```
