```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np

nii_file = '/Users/divya/Desktop/Pancreatic_Tumor/pancreas_001.nii.gz'
img = nib.load(nii_file)

# img.shape

image_data = img.get_fdata()

plt.imshow(image_data[:, :, 10], cmap='gray')
plt.show()
print(image_data[:, :, 10])
```

```python
nii_label_file = '/Users/divya/Desktop/Pancreatic_Tumor/pancreas_001_label.nii.gz'
img_label = nib.load(nii_label_file)

image_data_label = img_label.get_fdata()

slices = img_label.shape[2]
print(slices)

best = 0
slice_no = 0
for x in range(slices):
    mat = np.matrix(image_data_label[:, :, x])
    pancreas = np.count_nonzero(mat==2)
    if pancreas > best:
        best = pancreas
        slice_no = x

print(slice_no)
```

```python
plt.imshow(image_data_label[:, :, slice_no], cmap='gray')
plt.show()
print(image_data_label[:, :, slice_no])
```
